```python

#env
from dotenv import load_dotenv
import os
load_dotenv()
grok_key=os.getenv("grok_key")
```

```python
#groq
from langchain_groq import ChatGroq
llm=ChatGroq(temperature=0,groq_api_key=grok_key,model_name="llama-3.1-70b-versatile")
response=llm.invoke("The first person to land on moon was ...")
response.content
```

```python
#pdf loader
from langchain.document_loaders import PyPDFDirectoryLoader
doc_pdf=PyPDFDirectoryLoader('./docs').load()
doc_pdf
```

```python
#chunk
from langchain.text_splitter import RecursiveCharacterTextSplitter
chunk=RecursiveCharacterTextSplitter(chunk_size=800,chunk_overlap=50).split_documents(doc_pdf)
chunk
```

```python
#file txt read
import os
folder_path="./docs"
file_list=[]
for file_name in os.listdir(folder_path):
   if file_name.endswith(".txt"):
      with open(os.path.join(folder_path,file_name), 'r') as file:
         file_list.append({"name":file_name,"data":file.read()})
file_list
```

```python
#chromadb
import chromadb
chromadb_collection=chromadb.Client().get_or_create_collection(name="my_collection")

documents=["This is a document about cat", "This is a document about car", "This is a document about bike"]
ids=["id1", "id2","id3"]
metadatas=[{"category": "animal"}, {"category": "vehicle"}, {"category": "vehicle"}]

chromadb_collection.add(documents=documents,ids=ids,metadatas=metadatas)

result=chromadb_collection.query(query_texts=["vehicle"],n_results=1)
result
```
