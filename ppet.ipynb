```python
#file txt read
import os
folder_path="./docs"
file_list=[]
for file_name in os.listdir(folder_path):
   if file_name.endswith(".txt"):
      with open(os.path.join(folder_path,file_name), 'r') as file:
         file_list.append({"name":file_name,"data":file.read()})
file_list
```

```python
#chromadb
import chromadb
chromadb_collection_1=chromadb.Client().get_or_create_collection(name="pet1")
chromadb_collection_2=chromadb.Client().get_or_create_collection(name="pet2")

#sentence transformers
from sentence_transformers import SentenceTransformer
model=SentenceTransformer('paraphrase-MiniLM-L3-v2')

#data prepare
documents,metadatas,ids,embeddings=[],[],[],[]
for index,file in enumerate(file_list):
    documents.append(file['data'])
    metadatas.append({'source':file['name']})
    ids.append(str(index + 1))
    embedding=model.encode(file['data']).tolist()
    embeddings.append(embedding)

#data add
chromadb_collection_1.add(documents=documents,ids=ids,metadatas=metadatas)
chromadb_collection_2.add(documents=documents,ids=ids,metadatas=metadatas,embeddings=embeddings)

#search
results=chromadb_collection_1.query(query_texts=["What is the Nutrition needs of the pet animals?"],n_results=1)
results

input_em=model.encode("What are the different kinds of pets people commonly own?").tolist()
results=chromadb_collection_2.query(query_embeddings=[input_em],n_results=1)
results
```
